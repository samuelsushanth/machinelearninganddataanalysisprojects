```python
# Count the number of .zip files
zip_count = df['submission'].str.endswith('.zip').sum()

# Count the total number of submissions
total_submissions = len(df)

# Calculate the percentage of .zip submissions
zip_percentage = (zip_count / total_submissions) * 100

# Print the result
print(f"Percentage of submissions in .zip format: {zip_percentage:.2f}%")
```

```python
# Convert submit_time to datetime format
df['submit_time'] = pd.to_datetime(df['submit_time'], format='%m/%d/%y-%H:%M:%S')

# Define the first deadline (let's assume it's '01/03/17-20:00:00')
first_deadline = pd.to_datetime('01/03/17-23:59:59', format='%m/%d/%y-%H:%M:%S')
print(first_deadline)

# Filter out students who submitted after the first deadline
submissions_after_first_deadline = df[df['submit_time'] >= first_deadline]
print(submissions_after_first_deadline)

# Count the number of students who submitted after the first deadline
count_after_first_deadline = len(submissions_after_first_deadline)
print(count_after_first_deadline)

print(f"Number of students who submitted after the first deadline: {count_after_first_deadline}")
```

```python
# prompt: group by only '%m/%d/%y in submit time and check the len of all submit dates

# Group by the date part of 'submit_time' and get the count of submissions for each date
submission_counts_by_date = df.groupby(df['submit_time'].dt.strftime('%m/%d/%y'))['submit_time'].count()

submission_counts_by_date
```

```python
# prompt: group by only %m/%d/%y-%H in submit time and check the len of all submit dates

# Group by the date and hour part of 'submit_time' and get the count of submissions for each date and hour
submission_counts_by_date_hour = df.groupby(df['submit_time'].dt.strftime('%m/%d/%y-%H'))['submit_time'].count()

print(submission_counts_by_date_hour)

print(len(submission_counts_by_date_hour))
```
