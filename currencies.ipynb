```python
import pandas as pd

# Load your dataset (assuming it's already in df)

# Convert 'Date' column to datetime format
df.rename(columns={'Currency': 'Date'}, inplace=True)  # Rename the first column to "Date" if needed
df['Date'] = pd.to_datetime(df['Date'])

# Convert all other columns to numeric values
df_numeric = df.drop(columns=['Date']).replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')
print(df_numeric)

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()
print(correlation_matrix)

# Find the correlation between Indian Rupee and other currencies
indian_rupee_correlation = correlation_matrix['Indian Rupee']
print(indian_rupee_correlation)

# Find the currency with the highest correlation (excluding itself)
most_correlated_currency = indian_rupee_correlation.dropna().drop('Indian Rupee').abs().idxmax()
print(most_correlated_currency)

# Print the most correlated currency
print(f"The Indian Rupee is most correlated with the {most_correlated_currency}")
```

```python
# Standardizing the currency names to match the index
selected_currencies = ['Australian Dollar', 'U.S. Dollar', 'Japanese Yen', 'Euro']

# Select the four rows in the correlation matrix for Indian Rupee
selected_correlations = indian_rupee_correlation.loc[selected_currencies]

# Display the selected correlations
selected_correlations

```

```python
# Define the list of currencies correctly
selected_currencies = ['Indian Rupee', 'Japanese Yen', 'U.S. Dollar', 'Euro',
                       'Australian Dollar', 'U.K. Pound Sterling', 'Chinese Yuan']

# Use .loc[] to select multiple columns
seven_currency_correlation = correlation_matrix.loc[selected_currencies, selected_currencies]

# Print the result
print(seven_currency_correlation)

```

```python
import numpy as np

# Replace the diagonal values with NaN to ignore self-correlation
np.fill_diagonal(seven_currency_correlation.values, np.nan)

# Find the minimum correlation value (most negative)
min_correlation = seven_currency_correlation.min().min()
print(min_correlation)

# Find the currency pair corresponding to the minimum correlation
min_pair = seven_currency_correlation.stack().idxmin()
print(min_pair)

print(f"Highest negative correlation is between {min_pair[0]} and {min_pair[1]} with a value of {min_correlation}")
```
