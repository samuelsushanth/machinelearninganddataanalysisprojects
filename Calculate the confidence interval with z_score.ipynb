```python
# prompt: standard deviation with a probablility distribution function graph

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.express as px
import warnings
# import necessary libraries


warnings.filterwarnings('ignore')
# read the dataset

#df=pd.read_excel('Inferential Statistics - Student Loan.xlsx')

df=pd.read_csv('Inferential Statistics - Powai Flats Rent.csv')
df.info()
df.describe()
df.head(10)
df.tail(10)
print(df.columns)

# Calculate the sum of the 'Monthly Rent' column
sum_monthly_rent = df['Monthly Rent'].sum()

# Calculate the total length of the 'Monthly Rent' column
total_length = len(df['Monthly Rent'])

# Calculate the average monthly rent
average_monthly_rent = sum_monthly_rent / total_length

print(f"Average Monthly Rent: {average_monthly_rent}")


# Calculate the standard deviation of the 'Monthly Rent' column
std_dev_monthly_rent = np.std(df['Monthly Rent'])
print(f"Standard Deviation of Monthly Rent: {std_dev_monthly_rent}")


# Create a probability distribution function (PDF) graph
plt.figure(figsize=(10, 6))
sb.histplot(df['Monthly Rent'], kde=True)
plt.title('Probability Distribution of Monthly Rent')
plt.xlabel('Monthly Rent')
plt.ylabel('Probability Density')
plt.show()

```

```python
# prompt: What is the standard deviation of 'Monthly Rent' ?

# Calculate the standard deviation of the 'Monthly Rent' column
std_dev_monthly_rent = np.std(df['Monthly Rent'])
print(f"Standard Deviation of Monthly Rent: {std_dev_monthly_rent}")
```

```python
# Calculate the total length of the 'Monthly Rent' column
total_length = len(df['Monthly Rent'])
print(f"Total Length of 'Monthly Rent' Column: {total_length}")
```

```python
# prompt: with  90% confidence interval calculate the z score

import scipy.stats as stats

# Assuming 'average_monthly_rent' and 'std_dev_monthly_rent' are already calculated
confidence_level = 0.90
z_score = stats.norm.ppf((1 + confidence_level) / 2)

print(f"Z-score for a {confidence_level * 100}% confidence interval: {z_score}")
```

```python
# prompt: Calculate the confidence interval with z_score at 90 % confidence level and a sample mean and standard deviation of column 'Monthly Rent'

# Assuming 'average_monthly_rent' and 'std_dev_monthly_rent' are already calculated
confidence_level = 0.90
z_score = stats.norm.ppf((1 + confidence_level) / 2)
n = len(df['Monthly Rent'])  # Sample size

margin_of_error = z_score * (std_dev_monthly_rent / np.sqrt(n))

lower_bound = average_monthly_rent - margin_of_error
upper_bound = average_monthly_rent + margin_of_error

print(f"Confidence Interval ({confidence_level * 100}%): ({lower_bound:.2f}, {upper_bound:.2f})")
```
