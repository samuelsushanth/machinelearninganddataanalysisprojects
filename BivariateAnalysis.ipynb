```python
# prompt: Bivariate Analysis
# What is the correlation in Gold and Silver prices (round off the answer to two decimal places)? Also, enter a fraction, not a percentage (e.g. 0.34, not 34.00%).

# Calculate the correlation between Gold and Silver prices
correlation = df['GoldPrice'].corr(df['SilverPrice'])

# Round off the correlation to two decimal places
rounded_correlation = round(correlation, 2)

# Print the rounded correlation
rounded_correlation
```
