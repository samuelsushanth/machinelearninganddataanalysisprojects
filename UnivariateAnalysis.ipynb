```python
#Runs Scored by Tendulkar
#Plot a bar chart showing runs scored on the x-axis and frequency/count on the y-axis. In which bucket has he scored runs the most often?
# Create bins for run ranges
bins = [0, 10, 20, 30, 40, 50, 100, 200, 300]
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-100', '101-200', '201-300']

# Create a new column with binned run scores
odi['Run_Buckets'] = pd.cut(odi['Runs'], bins=bins, labels=labels, include_lowest=True)

# Count the frequency of each bucket
run_frequency = odi['Run_Buckets'].value_counts().sort_index()

# Create the bar plot
plt.figure(figsize=(12, 6))
run_frequency.plot(kind='bar')
plt.title("Distribution of Sachin Tendulkar's Run Scores in ODIs")
plt.xlabel("Runs Scored")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print the bucket with the highest frequency
most_frequent_bucket = run_frequency.idxmax()
print(f"Tendulkar scored runs most often in the {most_frequent_bucket} runs bucket.")
print("\nFrequency distribution:")
print(run_frequency)
```
