```python
# prompt: drop last two cloumns 'Unnamed: 4', 'Unnamed: 5'

df = df.drop(['Unnamed: 4', 'Unnamed: 5'], axis=1)
print(df.head(10))
```

```python
# prompt: clip the % string in column 'Recovery (%)'

df['Recovery (%)'] = df['Recovery (%)'].str.rstrip('%').astype('float')
print(df.head(10))
```

```python
# prompt: calculate new column as title Expected_loss and calculation of columns as (Probability of Default*Exposure at Default (in lakh Rs.)*(1-(Recovery (%))) for each of the customer

df['Expected_loss'] = df['Probability of Default'] * df['Exposure at Default (in lakh Rs.)'] * (1 - (df['Recovery (%)'] / 100))
print(df.head(10))
```

```python
# prompt: sum of the column Expected_loss

expected_loss_sum = df['Expected_loss'].sum()
print(f"The sum of the 'Expected_loss' column is: {expected_loss_sum}")
```
