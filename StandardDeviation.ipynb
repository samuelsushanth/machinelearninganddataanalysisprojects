```python
# prompt: plot the standard deviation with a probablility distribution function graph

import plotly.figure_factory as ff

# Assuming 'df' is your DataFrame and 'Sample Mean' is the column
hist_data = [df['Sample Mean']]
group_labels = ['Sample Mean']

fig = ff.create_distplot(hist_data, group_labels, bin_size=.2)
fig.show()
```

```python
# prompt: What is the standard deviation of our sampling distribution?

# Calculate the standard deviation of the sampling distribution (standard error)
standard_deviation_of_sampling_distribution = df['Sample Mean'].std()

print(f"The standard deviation of the sampling distribution is: {standard_deviation_of_sampling_distribution}")
```

```python
# prompt: Using the mean from the previous question, you can find the standard deviation using the formula -

# . Note that you would divide by n and not n-1 as you have the data for all 100 entries of the distribution and you don't need to sample the distribution.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.express as px
import warnings
import plotly.figure_factory as ff

warnings.filterwarnings('ignore')

# Assuming 'df' is your DataFrame and 'Sample Mean' is the column
# ... (your existing code for reading data and calculating the mean)

# Calculate the standard deviation of the sampling distribution
# Using the provided formula: standard_deviation^2 = Σ(Xi - X̄)^2 / n
sample_means = df['Sample Mean']
n = len(sample_means)
squared_differences_sum = np.sum((sample_means - sample_means.mean())**2)
standard_deviation = np.sqrt(squared_differences_sum / n)

print(f"The standard deviation of the sampling distribution (calculated using the provided formula) is: {standard_deviation}")
```
